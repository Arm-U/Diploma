```python
def preprocess_text_with_slash(text):
    if not "/" in text:
        return text
    # Split the text into phrases by slashes
    if ' / ' in text:
        phrases = text.split(' / ')
    elif "/" in text:
        phrases = text.split('/')
    
    # Join the phrases with parentheses around the second and subsequent phrases
    preprocessed_text = phrases[0] + ' (' + ','.join(phrases[1:]) + ')'
    
    return preprocessed_text

# # Example text
# text = "olla matkalla / reissussa"
# preprocessed_text = preprocess_text_with_slash(text)
# preprocessed_text
```

```python
import re


def preprocess_line(line):
    # Skip empty lines
    
    if not line.strip():
        return None
    line = line.replace(" + maan/mään", "")
    line = line.replace(" + maan/mään", "")
    line = line.replace(" + na/nä", "")
    line = line.replace("nimetä (1) + G + mukaan", "nimetä ... mukaan")
    

    
    
    line = line.replace("Huom! ", "")
    line = line.replace(" + ksi", "")
    line = line.replace(" +ksi", "")
    line = line.replace(" + P tai maan/mään", "")
    
    line = line.replace("*", "")
    
    line = line.replace(" (prosessi)", "")
    line = line.replace(" (negatiivinen)", "")
    line = line.replace(" (ainesanat)", "")
    line = line.replace(" (verbi + P)", "")
    line = line.replace(" (mon.)", "")
    if line.startswith("MIHIN + "):
        line = line.replace("MIHIN + ", "")
    if line.startswith("ILL + "):
        line = line.replace("ILL + ", "")
    if line.startswith("STA/STÄ + "):
        line = line.replace("STA/STÄ + ", "")
    if line.startswith("sta/stä + "):
        line = line.replace("sta/stä + ", "")
    if line.startswith("lle + "):
        line = line.replace("lle + ", "")
    line = line.replace(" + sta (stä)", "")
    line = line.replace(" + sta/stä", "")
    line = line.replace("+ sta/stä", "")
    line = line.replace(" + maan (mään)", "")
    line = line.replace(" + mastä/mästä", "")
    
    line = line.replace(" + P + na/nä", "")
    
    
    line = line.replace("ei + kahteen vuoteen", "ei ... kahteen vuoteen")
    
    line = line.replace(" + kanssa", "")
    
    # toto check
    line = line.replace(" + Ita/ltä", "")
    line = line.replace(" + lta/ltä", "")
    line = line.replace(" + na/nä", "")
    
    
    
    
        

    if line.startswith("-") and len(line) < 10:
        return None
    # remove " (1)" from the words
    line = re.sub(r' \(\d\)', "", line)
    line = line.replace(" + MIHIN", "")
    line = line.replace(" + MISTÄ", "")
    line = line.replace(" + S-MIHIN", "")
    line = line.replace(" + MISSÄ", "")
    line = line.replace(" + lle", "")
    

    
    # ad-hoc fixes
    # b'\xef\xb8\x8e sairas (kipe\xc3\xa4)'
    # replace doesn't work for bytes, so we need ad-hoc fixe
#     if 'sairas' in line:
#         print(line.encode())
    bug_char = b'\xef\xb8\x8e'
    bug_str = bug_char.decode('utf-8')
    if bug_str in line:
        #print(line.encode())
        line = line.replace(bug_str, '')
#        print(line.encode())

    if line == "Huom.":
        return None
    
    line = line.strip()

    # skip (Verbit: lähtea, mennä, tulla...)
    if line.startswith("("):
        return None
    
    # Generalize replacement for patterns like "Letter + word" and "Letter + word/word"
    # Capture groups: Letter, first word, optional second word
    line = re.sub(r'\b([A-Za-z]) \+ ([^\s/]+)(?:/([^\s/]+))?', lambda m: f"{m.group(2)} ({m.group(3)})" if m.group(3) else m.group(2), line)


    # Skip "Itä/Itä" like patterns
    if re.match(r'^[A-Za-zäöåÄÖÅ]{1,3}/[A-Za-zäöåÄÖÅ]{1,3}$', line):
        return None

    # Specific handling for "eräänä päivänä (= yhtenä päivänä)"
    if re.search(r'\(= ', line):
        line = re.sub(r' \(= ', ' (', line)
    else:
        # Handle specific complex cases first to ensure precise control over replacements
        complex_pattern_match = re.search(r'(.*) \(\d+\) \+ [A-Z]+ = (.*)', line)
        if complex_pattern_match:
            # Extract parts before and after "="
            first_part, second_part = complex_pattern_match.groups()
            # Reformat line according to the rule
            line = f"{first_part} ({second_part})"
        else:
            # If the word is like "kirjoittaa (1) + sta/stä" or "saapua (1)", keep everything before brackets
            line = re.sub(r' \(\d\).*', '', line)

            # If the word is like "ympäri + P" or "ympäri + MIHIN", keep the first part
            line = re.sub(r' \+ [A-Z]+$', '', line)

            # Handle "=" replacements more carefully
            if '=' in line:
                # For single "=" sign cases, replace and add closing parenthesis correctly
                if line.count('=') == 1:
                    line = re.sub(r' = ', ' (', line) + ")"
                # If there are several "=" signs
                elif line.count('=') > 1:
                    parts = line.split(' = ')
                    line = f"{parts[0]} ({', '.join(parts[1:])})"
                # Remove any trailing unmatched ")" that may have been added incorrectly
                line = re.sub(r'\)$', '', line) if line.count('(') < line.count(')') else line
                
    # slash shouldn't be in result, causing problems with audio file names 
    line = preprocess_text_with_slash(line)
    line = line.strip()
    if "+" in line:
        # warning!!!
        print("WARNINg!!")
        print(line)
    
    # Specific checks and replacements
    processed_lines = []  # To handle multiple output lines
    if '↔' in line:
        # Split the line for words like "aikaisin ↔ myöhään" into two separate lines
        parts = line.split('↔')
        processed_lines.extend([item.strip() for item in parts])
    else:
        processed_lines.append(line)

    # Return a list of processed lines
    return processed_lines

#     return line



def preprocess_file(input_file, output_file):
    with open(input_file, 'r', encoding='utf-8') as infile, open(output_file, 'w', encoding='utf-8') as outfile:
        for line in infile:
            processed_lines = preprocess_line(line.strip())
            if processed_lines is not None:
                for pline in processed_lines:
                    outfile.write(pline + '\n')

```

## Step 2.1 check for already existing words and remove them from new vocab

```python
import json
import os

def collect_existing_words(json_folder_path):
    existing_words = set()
    for file_name in os.listdir(json_folder_path):
        if file_name.endswith('.json'):
            file_path = os.path.join(json_folder_path, file_name)
            with open(file_path, 'r', encoding='utf-8') as file:
                data = json.load(file)
                for entry in data:
                    # Add the word in lowercase to the set
                    existing_words.add(entry["wordSecondLang"].lower().strip())
    return existing_words
```

```python
import csv

def filter_tsv_file(input_tsv_path, output_tsv_path, existing_words):
    seen_words_in_current_file = set()
    with open(input_tsv_path, 'r', encoding='utf-8') as infile, \
         open(output_tsv_path, 'w', newline='', encoding='utf-8') as outfile:
        reader = csv.reader(infile, delimiter='\t')
        writer = csv.writer(outfile, delimiter='\t')
        print(output_tsv_path)
        
        for row in reader:
            if len(row) == 0:
                continue
            if len(row) < 2:
                word = row[0]
                sentence = ""
            else:
                word, sentence = row
            word = word.strip()
            sentence = sentence.strip()
            
            # Check if the lowercase word is in the set of existing words
            if word.lower() not in existing_words and word.lower() not in seen_words_in_current_file:
                writer.writerow([word, sentence])
                seen_words_in_current_file.add(word.lower())
            else:
                if word.lower() in seen_words_in_current_file:
                    print(f"duplicate same file: {word}")        
                else:
                    print(f"duplicate previous files: {word}")        
```

```python
# Ensure word is a separate word or phrase in the phrase
def word_in_phrase(word, phrase, in_recursion = False):
    # fix cases like
    word = word.lower()
    phrase = phrase.lower()
    phrase = re.sub(r'[^\w\s]', '', phrase)
    if " " in word:
        word_parts = word.split()
        for word_part in word_parts:
            result = word_in_phrase(word_part, phrase, in_recursion=True)
            if result == False:
                return False
        print(word)
        print(phrase)
        return True
    if not word in phrase:
        return False
    else:
        if word in phrase.split():
            return True
        else:
            if not in_recursion:
                # check manually and debug
                print("=====Check manually")
                print(word)
                print(phrase)
                print("=====end of  manually")
            return False
        
# todo in chapter 5 it didn't work for some reason, double check
def process_file_4_phrase_examples(input_file, output_file):
    with open(input_file, 'r', encoding='utf-8') as infile:
        lines = [line.strip() for line in infile if line.strip()]

    # Separate words and phrases based on the updated criteria
    words = [line for line in lines if line[0].islower()]
    phrases = [line for line in lines if line[0].isupper() and ' ' in line]

    # Dictionary to keep track of used phrases
    used_phrases = {}

    # Find phrases containing each word and prepare output
    output_lines = []
    for line in lines:
        if line in words:
            phrase_found = False
            for phrase in phrases:
                if word_in_phrase(line, phrase) and phrase not in used_phrases:  
                    output_lines.append(f"{line}\t{phrase}")
                    phrase_found = True
                    used_phrases[phrase] = True
                    break  # Stop after finding the first phrase for each word
            if not phrase_found:
                output_lines.append(line)
        else:
            if not line in used_phrases:
                output_lines.append(line)
            

    # Write output to file
    with open(output_file, 'w', encoding='utf-8') as outfile:
        for line in output_lines:
            outfile.write(line + '\n')

# Paths to your input and output files
step2_2_output_file = folder + 'step2_2.txt'

process_file_4_phrase_examples(step2_1_output_file, step2_2_output_file)
```

## Step 3.1: postprocess sentence examples, remove """ and words in brackets

```python
import csv
import re

def process_file_step3(input_file_path, output_file_path):
    processed_data = []
    
    with open(input_file_path, 'r', encoding='utf-8') as infile:
        reader = csv.reader(infile, delimiter='\t')
        for row in reader:
            if len(row) != 2:
                print(f"Alert: should not skip row: {row}")
                continue  # Skip rows that don't have exactly 2 columns
            
            finnish_word, finnish_sentence = row
            
            # Remove all occurrences of '""""'
#             finnish_sentence = finnish_sentence.replace('"""', '')
#             print(finnish_sentence)
            if len(finnish_sentence) > 1:
                if finnish_sentence[0] == '"' and finnish_sentence[-1] == '"':
                    finnish_sentence = finnish_sentence.replace('"', '')
#             print(finnish_sentence)
            
            # Remove words in brackets (including the brackets) and any trailing space after them
            # This regex also accounts for nested brackets
            finnish_sentence = re.sub(r'\(.*?\)\s*', '', finnish_sentence)
            
            processed_data.append([finnish_word, finnish_sentence])
    
    with open(output_file_path, 'w', newline='', encoding='utf-8') as outfile:
        writer = csv.writer(outfile, delimiter='\t')
        writer.writerows(processed_data)

# Specify your input and output file paths
# output_file_path_step3_1 = 'sm2newch1_sanasto_processed_gpt_sentences_step3_1.txt'
step3_1_output_file = folder + 'step3_1.txt'

# Process the file
process_file_step3(step3_output_file, step3_1_output_file)
# process_file_step3("tmp.txt", output_file_path_step3_1)
```

## Step 4: Prepare translations to russian

```python
import csv

def read_input_tsv_step4(file_path):
    data = []
    with open(file_path, 'r', encoding='utf-8') as file:
        reader = csv.reader(file, delimiter='\t')
        for row in reader:
            if len(row) == 2:
                data.append(row)
            else:
                # Handle rows with only one column
                data.append([row[0], ''])
    return data

def write_output_tsv_step4(file_path, data):
    with open(file_path, 'w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file, delimiter='\t')
        for row in data:
            writer.writerow(row)

```

```python
step4_output_file = folder + 'step4.txt'

# Read the input data
input_data = read_input_tsv_step4(step3_1_output_file)

# Translate the data to Russian
# translated_data = gptutils.translate_to_russian(input_data, model = GPT_4)
translated_data = gptutils.translate_to_russian(input_data, model = DEFAULT_MODEL)
# translated_data = gptutils.translate_to_russian(input_data, model = GPT_3_5_TURBO)

# Write the output file
write_output_tsv_step4(step4_output_file, translated_data)
```

## Step 5: Clean gpt output

```python
def remove_pattern(text):
    # Split the text into lines
    lines = text.splitlines()
    
    # Filter out lines that contain only '==='
    filtered_lines = [line for line in lines if line.strip() != '===']
    
    # Join the remaining lines back into a single string with new lines
    clean_text = " ".join(filtered_lines)
    
    return clean_text

# new bug found GPTo
def fix_word(text):
    if ('===') in text:
        print("##start##")
        print(text)
        print("##end##")
        text = remove_pattern(text)
        print(text)
    return text
    

def read_and_postprocess_gpt_output_step5(file_path):
    processed_data = []
    seen_words = set()
    
    # Mapping of stressed vowels to their unstressed counterparts
    stress_replacements = {
        'а́': 'а', 'е́': 'е', 'и́': 'и', 'о́': 'о', 'у́': 'у',
        'ы́': 'ы', 'э́': 'э', 'ю́': 'ю', 'я́': 'я',
    }

    with open(file_path, 'r', encoding='utf-8') as file:
        reader = csv.reader(file, delimiter='\t')
        for row in reader:
            if len(row) != 4:  # Ensure row has exactly 4 columns
                continue

            finnish_word, finnish_sentence, russian_word, russian_sentence = row
            
            # Skip duplicates based on the first column
            if finnish_word in seen_words:
                continue
            seen_words.add(finnish_word)
            
            # Apply transformations as before, including stress removal
            if finnish_word[0].islower():
                if russian_word[0].isupper():
                    russian_word = russian_word[0].lower() + russian_word[1:]
                if russian_word.endswith("."):
                    russian_word = russian_word[:-1]
            
            for stressed, unstressed in stress_replacements.items():
                russian_word = russian_word.replace(stressed, unstressed)
                russian_sentence = russian_sentence.replace(stressed, unstressed)
            
            russian_word = russian_word.replace("'", "").replace("#", "")
            russian_sentence = russian_sentence.replace("'", "").replace("#", "")
            
            if "(" not in finnish_sentence and "(" in russian_sentence and ")" in russian_sentence:
                start = russian_sentence.find("(")
                end = russian_sentence.find(")") + 1
                russian_sentence = russian_sentence[:start] + russian_sentence[end:]
                
            if len(russian_word) > 1:
                if russian_word[0] == '"' and russian_word[-1] == '"':
                    russian_word = russian_word.replace('"', '')
                    
            if len(russian_sentence) > 1:
                if russian_sentence[0] == '"' and russian_sentence[-1] == '"':
                    russian_sentence = russian_sentence.replace('"', '')
                    
            if finnish_word.endswith(" ksi"):
                finnish_word.replace(" ksi", "")
                

                
            russian_word = russian_word.split("\n")[0]
            russian_sentence = fix_word(russian_sentence)
            russian_word = fix_word(russian_word)
            
            processed_data.append([finnish_word, finnish_sentence, russian_word, russian_sentence])
    
    return processed_data
```

```python
processed_data = read_and_postprocess_gpt_output_step5(step4_output_file)
```
